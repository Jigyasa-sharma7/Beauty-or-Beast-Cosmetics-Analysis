```python
# percentage of missing values in each columns
import pandas as pd
data=pd.read_csv('cscpopendata.csv')
missing_values_percentage = data.isnull().mean() * 100
missing_columns = missing_values_percentage[missing_values_percentage > 0]
for column, percentage in missing_columns.items():
    print(f"missing values in {column}: {percentage:.2f}%")
```

the logic behind it is:
isnull convert each missing or NAn value as 1 and non-null value as 0 .
so after computing the mean we take 1+0(true or false values)/number of rows 
and after * 100 we get the percentage of missing values in each column. item() is a iterative loop for dict and panda series it prints both key and value

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data=pd.read_csv('cscpopendata.csv')
# Handling missing values
columns_to_drop = ['DiscontinuedDate', 'ChemicalDateRemoved']
data_cleaned = data.drop(columns=columns_to_drop)

# Fill missing values for specific columns
data_cleaned['BrandName'] = data_cleaned['BrandName'].fillna('Unknown Brand')
data_cleaned['CasNumber'] = data_cleaned['CasNumber'].fillna('Unknown CasNumber')

# Parsing Date Fields
date_columns = ['InitialDateReported', 'MostRecentDateReported', 'ChemicalCreatedAt', 'ChemicalUpdatedAt']
for col in date_columns:
    data_cleaned[col] = pd.to_datetime(data_cleaned[col], errors='coerce')

# Formatting Categorical Variables
categorical_columns = ['PrimaryCategory', 'SubCategory', 'BrandName', 'CompanyName']
for col in categorical_columns:
    data_cleaned[col] = data_cleaned[col].str.strip().str.title()  # Remove extra spaces and standardize case

# Clean text in `ProductName`
data_cleaned['ProductName'] = data_cleaned['ProductName'].str.replace(r'[^a-zA-Z0-9\s-]', '', regex=True)

# delete the duplicates
data.drop_duplicates()

# Visualizing Top Categories
top_categories = data_cleaned['PrimaryCategory'].value_counts()
# Plot the distribution of top categories
plt.figure(figsize=(10, 6))
sns.barplot(y=top_categories.index, x=top_categories.values, palette="viridis")
plt.title("Top 10 Primary Categories by Product Count", fontsize=16)
plt.xlabel("Number of Products")
plt.ylabel("Primary Category")
plt.tight_layout()
plt.show()
```

```python
# Trends Over Time
data_cleaned['YearReported'] = data_cleaned['InitialDateReported'].dt.year

# Count of products reported each year
yearly_trends = data_cleaned['YearReported'].value_counts().sort_index()

# Plot the trend over years
plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_trends.index, y=yearly_trends.values, marker="o", color="b")
plt.title("Number of Products Reported Over the Years", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Number of Products")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#  Most Common Chemicals
top_chemicals = data_cleaned['ChemicalName'].value_counts().head(10)

# Plot the top chemicals
plt.figure(figsize=(10, 6))
sns.barplot(y=top_chemicals.index, x=top_chemicals.values, palette="magma")
plt.title("Top 10 Most Common Chemicals", fontsize=16)
plt.xlabel("Number of Products")
plt.ylabel("Chemical Name")
plt.tight_layout()
plt.show()
```
